```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import datetime
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
import json

%matplotlib inline
```

```python
# Load .env enviroment variables
load_dotenv()
```

```python
# Set Alpaca API key and secret
# YOUR CODE HERE!
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Verify that Alpaca key and secret were correctly loaded
print(f"Alpaca Key type: {type(alpaca_api_key)}")
print(f"Alpaca Secret Key type: {type(alpaca_secret_key)}")
```

```python
# Create the Alpaca API object
# YOUR CODE HERE!
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2020-02-01', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2022-03-05', tz='America/New_York').isoformat()
```

## TSLA Stock

```python
#TSLA Stock 
timeframe = '1Day'
tickers = ["TSLA"]

TSLA_df = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date
).df
```

```python
#Remove timestamp
TSLA_df.index.date
TSLA_df.index = TSLA_df.index.date
TSLA_df
```

```python
TSLA_df.to_csv('../Resources/TSLA_20-22.csv')
```

```python
#ticker symbol only
tsla_symbol = TSLA_df['symbol']
```

```python
#Isolate close price
tsla_df_close_price = TSLA_df['close']

```

```python
#calculate PCT
tsla_df_close_PCT =tsla_df_close_price.pct_change()
tsla_df_close_PCT
```

```python
#concat/export TSLA Dataframe
TSLA_df = pd.concat([tsla_symbol, tsla_df_close_PCT], axis="columns", join="inner")
TSLA_df.rename(columns={'close':'PCT_CLOSE'}, inplace=True)
TSLA_df.index.rename('Date', inplace=True)
TSLA_df.to_csv('../Resources/symbol_TSLA_PCT_close.csv')
TSLA_df
```
